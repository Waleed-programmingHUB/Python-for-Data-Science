## Barplots - Categorical Variable

````python
# barplots
# modules - libraries
import seaborn as _seaBorn
import matplotlib.pyplot as _matplot

# data visualization
_seaBorn.set_theme(style="ticks", color_codes=True)

titanic = _seaBorn.load_dataset("titanic")

_seaBorn.catplot(x ="sex", y ="survived", hue ="class", kind ="bar", data = titanic)
# show the data
_matplot.show()
````

## Countplots - Categorical Variable

````python
# Countplots
import seaborn as _seaborn
import matplotlib.pyplot as _matplot

_seaborn.set_theme(style='ticks', color_codes=True)
_titanic = _seaborn.load_dataset('titanic')
_mat1 = _seaborn.countplot(x = 'sex', data = _titanic, hue = 'class')
_mat1.set_title('Plot for Counting')
_matplot.show()
````

## Scatterplot - Continuous Variables

````python
# Scatterplot
import seaborn as _seaborn_scatter
import matplotlib.pyplot as _matplot_scatter

_seaborn_scatter.set_theme(style ="ticks", color_codes= True)
_scatter_titanic = _seaborn_scatter.load_dataset("titanic")
_scatterplot = _seaborn_scatter.FacetGrid(_scatter_titanic, row ='sex', hue ='alone')
_scatterplot = (_scatterplot.map(_matplot_scatter.scatter, 'age', 'fare').add_legend())

_matplot_scatter.show()
````

# Practice data set

````python
# using a built in dataset - planets
import seaborn as s_planets
import matplotlib.pyplot as _mat_planets

s_planets.set_theme(style='ticks', color_codes=True)
data_planet = s_planets.load_dataset('planets')
_s = s_planets.scatterplot(x='year', y='distance',hue='method', data=data_planet)
_s.set_title("Planets | Data Visualization")
_mat_planets.show()

````

# Using Histroplot | Data Visualization |Penguins Dateset 

````python
import seaborn as s_penguins
import matplotlib.pyplot as mat_pen

s_penguins.set_theme(style='dark',color_codes=True)

penguins = s_penguins.load_dataset("penguins")
s = s_penguins.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
s.set_title("Penguins in Atlantic Ocean 2022")
mat_pen.show()
````
