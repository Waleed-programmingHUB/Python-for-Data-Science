# Barplot Data Visualization

## Barplot 1

```python
# learning a Barplot
import seaborn as my_seaborn
import matplotlib.pyplot as my_mat

# theme
my_seaborn.set_theme(style='ticks',color_codes= True)
# data set
mydata = my_seaborn.load_dataset('iris') 
# plot graph
my_mat.figure(figsize=(18,8))
graph = my_seaborn.barplot(x='sepal_length',y='sepal_width',data= mydata)
# graph title 
graph.set_title('BarPlot | Iris Dataset | Data Visualization')
print(mydata)
# figure size

# show graph
my_mat.show()
```

## Barplot 3

```python
# learning a Barplot
import seaborn as my_seaborn
import matplotlib.pyplot as my_mat

# theme
my_seaborn.set_theme(style='ticks',color_codes= True)
# data set
mydata = my_seaborn.load_dataset('titanic') 
# figure size
my_mat.figure(figsize=(18,10))
# plot graph
graph = my_seaborn.barplot(x='sex',y='alone',hue='who',data= mydata)
# graph title 
graph.set_title('barplot | titanic dataset | data visualization'.upper())

# show graph
my_mat.show()
```

## Barplot 4 | data by order

```python
# learning a Barplot
import seaborn as my_seaborn
import matplotlib.pyplot as my_mat

# theme
my_seaborn.set_theme(style='ticks',color_codes= True)
# data set
mydata = my_seaborn.load_dataset('titanic') 
# plot graph
my_mat.figure(figsize=(18,8))
graph = my_seaborn.barplot(x='sex',y='alone',hue='who',data= mydata,order=['male','female'])
# graph title 
graph.set_title('barplot | titanic dataset | data visualization'.capitalize())
# figure size

# show graph
my_mat.show() 
```

## Barplot | Different Color pallete

```python
# learning a Barplot
import seaborn as my_seaborn
import matplotlib.pyplot as my_mat

# theme
my_seaborn.set_theme(style='ticks',color_codes= True)
# data set
mydata = my_seaborn.load_dataset('titanic') 
# figure size
my_mat.figure(figsize=(18,8))
# plot graph
graph = my_seaborn.barplot(
    x='sex', # x -axis
    y='alone',# y -axis
    hue='who',# presenting the element
    data= mydata, # data set
    color='#656565 ', # color bar
    order=['male','female']
    )
# graph title 
graph.set_title('barplot | titanic dataset | data visualization'.upper())

# show graph
my_mat.show() 
```
