# Barplot Data Visualization

## Barplot 1 | Intro

```python
# learning a Barplot
import seaborn as my_seaborn
import matplotlib.pyplot as my_mat

# theme
my_seaborn.set_theme(style='ticks',color_codes= True)
# data set
mydata = my_seaborn.load_dataset('iris') 
# plot graph
my_mat.figure(figsize=(18,8))
graph = my_seaborn.barplot(x='sepal_length',y='sepal_width',data= mydata ,ci=None)
# graph title 
graph.set_title('BarPlot | Iris Dataset | Data Visualization')
print(mydata)
# figure size

# show graph
my_mat.show()
```

## Barplot 7 | estimator | library using numpy | Palette | Estimator | Horizontal plot

```python
# learning a Barplot
import seaborn as my_seaborn
import matplotlib.pyplot as my_mat
import numpy as my_numpy

# theme
my_seaborn.set_theme(style='ticks',color_codes= True)
# data set
mydata = my_seaborn.load_dataset('titanic') 
# figure size
my_mat.figure(figsize=(18,8))
# plot graph
graph = my_seaborn.barplot(x='fare', y='class',hue='sex', data= mydata, color='#999933' ,ci=None ,estimator= my_numpy.median)
# graph title 
graph.set_title('barplot | titanic dataset | data visualization'.upper())

# show graph
my_mat.show() 
```

# Barplot 8 | estimator | library using numpy | Estimator | Saturation

```python
# learning a Barplot
import seaborn as my_seaborn
import matplotlib.pyplot as my_mat
import numpy as my_numpy

# theme
my_seaborn.set_theme(style='ticks',color_codes= True)
# data set
mydata = my_seaborn.load_dataset('titanic') 
# figure size
my_mat.figure(figsize=(15,8))
# plot graph
graph = my_seaborn.barplot(
    x='class', # x - axis
    y='fare', # y - axis
    hue='sex', # presenting a element
    data= mydata, # data set
    color='#565002', # bar-color background 
    ci=None , # error bars | confidence interval
    estimator= my_numpy.median, # estimating the data
    saturation=0.23)
# graph title 
graph.set_title('barplot | titanic dataset | data visualization'.upper())

# show graph
my_mat.show() 
```

# Barplot | Background Styling in Graph Plot

```python
# learning a Barplot
import seaborn as my_seaborn
import matplotlib.pyplot as my_mat
import numpy as my_numpy

# theme
my_seaborn.set_theme(style='ticks',color_codes= True)
# data set
mydata = my_seaborn.load_dataset('titanic') 
# figure size
my_mat.figure(figsize=(16,8))
# graph title 
my_mat.title('barplot | titanic dataset | data visualization'.upper())

# plot graph
graph = my_seaborn.barplot(
    x='class', # x - axis
    y='fare', # y - axis
    data= mydata, # data set
    estimator= my_numpy.median,
    linewidth=5,
    facecolor=(0.5,0.2,0.1,0.6),
    errcolor='.5',
    edgecolor='.2',
    ci=None)
# show graph
my_mat.show() 
```
